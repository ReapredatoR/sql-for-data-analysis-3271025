```python
#Import the libraries needed to run the code
import pandas as pd
import sqlalchemy as sa
import plotly.express as px
import plotly.graph_objects as go
from sqlalchemy.engine import create_engine
```

```python
# Connecting to the HPlusSport database
engine = create_engine("mysql+pymysql://mariadb:mariadb@localhost/hplussport?charset=utf8mb4")
connection = engine.connect()
```

```python
#Let's take a peak at our data. This code shows the first 5 records
df = pd.read_sql_query("SELECT * FROM Orders", connection)
df.head()
```

```python
#Now we'll make a simple bar chart of sales by year
fig = px.bar(df, x='CreationDate', y='TotalDue')
fig.show()
```

```python
import dash
from dash import dcc
from dash import html
import plotly.graph_objects as go

# Assuming 'SalespersonID' is the correct column name
fig = go.Figure()

for index, row in df.iterrows():
    df_salesperson = df[df['SalespersonID'] == row.SalespersonID]  # Change 'Salesperson' to 'SalespersonID'
    fig.add_trace(
        go.Scatter(x=list(df_salesperson.CreationDate),
                   y=list(df_salesperson['TotalDue']),
                   name=row.SalespersonID,  # Change 'Salesperson' to 'SalespersonID'
                   text=list(df['TotalDue'].round(1)),
                   textposition="top left"))

salesperson_list = []
visibility = [False for _ in range(len(df['SalespersonID']))]
for index, row in df.iterrows():
    visibility = [False for _ in range(len(df['SalespersonID']))]
    visibility[index] = True
    d = dict(label=row.SalespersonID,  # Change 'Salesperson' to 'SalespersonID'
             method="update",
             args=[{"visible": visibility},
                   {"title": "OrderSummary",
                    "annotations": []}])

    salesperson_list.append(d)

fig.update_layout(
    updatemenus=[
        dict(
            active=0,
            buttons=salesperson_list,
        )
    ])

app = dash.Dash()
app.layout = html.Div([
    dcc.Graph(figure=fig)
])

app.run_server(debug=True, use_reloader=False)
```
